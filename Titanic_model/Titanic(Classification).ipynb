```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_RFR = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
model_RFR.fit(x_train,y_train)
```

```python
model_RFR.score(x_train,y_train)
```

```python
random_forest_list = model_RFR.predict(test_data).tolist()
```

```python
for i in range(len(random_forest_list)) :
    if random_forest_list[i] < 0.5:
        random_forest_list[i] = 0
    else:
        random_forest_list[i] = 1
```

```python
from sklearn.tree import DecisionTreeRegressor 
```

```python
model_DTR = DecisionTreeRegressor(random_state = 0)
```

```python
model_DTR.fit(x_train,y_train)
```

```python
model_DTR.score(x_train,y_train)
```

```python
decision_trees_list = model_DTR.predict(test_data)
```

```python
for i in range(len(decision_trees_list)) :
    if decision_trees_list[i] < 0.5:
        decision_trees_list[i] = 0
    else:
        decision_trees_list[i] = 1
```
